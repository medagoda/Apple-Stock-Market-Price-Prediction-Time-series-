```python
import pandas as pd
import matplotlib.pyplot as plt

# Dictionary Data
data = {
    "Maths": [45, 67, 56, 78, 50, 40],
    "Science": [56, 75, 65, 89, 60, 45],
    "English": [78, 88, 75, 90, 70, 60]
}

# Convert Dictionary to DataFrame
df = pd.DataFrame(data)

# Draw Boxplot
plt.figure(figsize=(8, 6))
plt.title("Student Marks Distribution")
plt.boxplot([df["Maths"], df["Science"], df["English"]],
            labels=["Maths", "Science", "English"],
            patch_artist=True,
            boxprops=dict(facecolor='lightgreen'))

plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dictionary Dataset
employee_data = {
    "Age": [22, 25, 28, 30, 35, 40, 45, 50, 55, 60],
    "Salary": [25000, 30000, 35000, 40000, 45000, 50000, 60000, 70000, 80000, 90000]
}

# Convert to DataFrame
df = pd.DataFrame(employee_data)

# Scatter Plot
plt.figure(figsize=(8, 6))
plt.title("Age vs Salary")
plt.scatter(df["Age"], df["Salary"], color='red', marker='o')

plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Dictionary Data
employee_data = {
    "Salary": [25000, 30000, 35000, 40000, 45000, 50000, 60000, 70000, 80000, 90000]
}

# Convert to DataFrame
df = pd.DataFrame(employee_data)

# Sort Salary Values
sorted_salary = np.sort(df["Salary"])

# Quantiles
quantiles = np.linspace(0, 1, len(sorted_salary))

# Quantile Plot
plt.figure(figsize=(8, 6))
plt.title("Quantile Plot of Salary")
plt.plot(quantiles, sorted_salary, color='green', marker='o', linestyle='dashed', markersize=8)
plt.xlabel("Quantiles")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate synthetic data from a normal distribution (mean = 0, std = 1)
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=0, scale=1, size=1000)

# Create Q-Q plot
stats.probplot(data, dist="norm", plot=plt)
plt.title("Q-Q Plot for Normally Distributed Data")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data from a normal distribution (mean = 0, std = 1)
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=0, scale=1, size=1000)

# Create a Histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)  # 30 bins for better granularity
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate synthetic data for demonstration
np.random.seed(42)

# Positive correlation (x increases, y increases)
x_positive = np.random.normal(0, 1, 100)
y_positive = 2 * x_positive + np.random.normal(0, 0.5, 100)

# Negative correlation (x increases, y decreases)
x_negative = np.random.normal(0, 1, 100)
y_negative = -2 * x_negative + np.random.normal(0, 0.5, 100)

# No correlation (random data)
x_no_corr = np.random.normal(0, 1, 100)
y_no_corr = np.random.normal(0, 1, 100)

# Calculate correlation coefficients
corr_pos = np.corrcoef(x_positive, y_positive)[0, 1]
corr_neg = np.corrcoef(x_negative, y_negative)[0, 1]
corr_no = np.corrcoef(x_no_corr, y_no_corr)[0, 1]

# Display correlation coefficients
print(f"Positive Correlation Coefficient: {corr_pos:.2f}")
print(f"Negative Correlation Coefficient: {corr_neg:.2f}")
print(f"No Correlation Coefficient: {corr_no:.2f}")

# Scatter plots for visualization
plt.figure(figsize=(12, 4))

# Positive correlation plot
plt.subplot(1, 3, 1)
plt.scatter(x_positive, y_positive, color='blue')
plt.title(f"Positive Correlation: {corr_pos:.2f}")
plt.xlabel("X")
plt.ylabel("Y")

# Negative correlation plot
plt.subplot(1, 3, 2)
plt.scatter(x_negative, y_negative, color='red')
plt.title(f"Negative Correlation: {corr_neg:.2f}")
plt.xlabel("X")
plt.ylabel("Y")

# No correlation plot
plt.subplot(1, 3, 3)
plt.scatter(x_no_corr, y_no_corr, color='green')
plt.title(f"No Correlation: {corr_no:.2f}")
plt.xlabel("X")
plt.ylabel("Y")

plt.tight_layout()
plt.show()
```
